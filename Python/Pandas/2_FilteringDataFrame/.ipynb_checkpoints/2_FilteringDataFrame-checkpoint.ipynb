# Filtering Data Frame

```python
import pandas as pd
```

```python
# Data taken from https://github.com/fivethirtyeight/data/tree/master/airline-safety
df = pd.read_csv('airline-safety.csv', index_col='airline')
```

```python
df.head()
```

```python
# Creating a boolean series to filter
df.fatalities_00_14 > 0
```

```python
df[df.fatalities_00_14 > 0]['fatalities_00_14'].sort_values(ascending=False)
```

```python
df[df.incidents_00_14 > 0][['incidents_00_14', 'fatal_accidents_00_14', 'fatalities_00_14']].sort_values(by='incidents_00_14', ascending=False)
```
