# Categorizing

Using groupby, pandas can also group dataframe based on a certain attribute. The same like in SQL.

```python
import pandas as pd
```

```python
di = pd.read_csv('./iris_dataset.csv')
di.head()
```

```python
di.groupby('flower').count()
```

```python
di.groupby('flower').mean()
```

```python
di.groupby('flower').std()
```

```python
di.groupby('flower').max()
```

Aggregation can also be done using .agg() with the list of aggregation function to call.


```python
di.groupby('flower').agg(['max', 'sum', 'std'])
```

```python
di.groupby('flower')[['sepal_width']].agg(['max', 'sum', 'std'])
```
