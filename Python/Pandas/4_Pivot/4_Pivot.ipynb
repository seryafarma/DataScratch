## Intro
Some explanation about Pivot can be found in: https://en.wikipedia.org/wiki/Pivot_table


Pivot table is useful when at least there are two classification in the dataset.
For example using wikipedia: https://en.wikipedia.org/wiki/Pivot_table
The question is: How many Units did we sell in each Region for every Ship Date?

Classification: Region, Date.
Value: Units.

In pandas, a real pivot() needs to have unique combinations of row and column.
however pivot_table() does not, it simply aggregates them.

```python
da = pd.read_csv('./school_grades_dataset.csv')
da.head()
```

```python
da.pivot_table(index='school', columns='sex', values='G1')
```

By default the pivot_table() returns the mean of the F and M in this case. If another aggregation is needed it can be done by passing aggfunc='' argument to the function.
